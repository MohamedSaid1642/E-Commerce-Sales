# Online Retail Data Pipeline

This notebook processes retail transaction data through Bronze → Silver → Golden layers and loads it into SQL Server for analysis.

**Stages:**
1. **Bronze Layer**: Raw data import and initial exploration
2. **Silver Layer**: Data cleaning and type conversion
3. **Golden Layer**: Final dataset split into Sales, Refunds, and Discounts tables
4. **Database**: Upload cleaned data to SQL Server

```python
# ============================================
# BRONZE LAYER: IMPORTS & RAW DATA LOADING
# ============================================

import pandas as pd
import numpy as np
```

```python
# Load raw retail transaction data from CSV
df_bronze = pd.read_csv('online_retail_II.csv')
```

```python
# Display raw dataset
df_bronze
```

```python
# Check data structure and data types
df_bronze.info()
```

```python
# Display statistical summary of numeric columns
df_bronze.describe()
```

## SILVER LAYER: FINAL CLEANED DATA

```python
# Create silver layer copy and optimize data types
df_silver = df_bronze
```

```python
# Convert to optimal pandas data types (nullable types)
df_silver = df_silver.convert_dtypes()
```

```python
# Verify optimized data types in silver layer
df_silver.info()
```

```python
# Display statistical summary of silver layer
df_silver.describe()
```

```python
# Display sample of cleaned data
df_silver.sample(5)
```

```python
# Remove rows with missing values and check updated structure
df_silver = df_silver.dropna()
df_silver.info()
```

```python
# Identify any remaining negative quantities in silver layer
df_silver[df_silver['Quantity'] <= 0]
```

## SILVER LAYER: FINAL CLEANED DATA

```python
# Create golden layer (final clean dataset)
df_golden = df_silver
df_golden
```

```python
# Verify cleaned data structure
df_golden.info()
```

## DATA SPLITTING: Categorize into Sales, Refunds, and Discounts

```python
# Extract sales transactions (positive quantity)
sales_df = df_golden[df_golden['Quantity'] > 0]
sales_df
```

```python
# Extract refund transactions (negative quantity)
refund_df = df_golden[df_golden['Quantity'] < 0]
refund_df
```

```python
# Convert negative quantities to positive for clarity in refunds table
refund_df['Quantity'] = refund_df['Quantity'] * -1
refund_df
```

```python
# Display refund data with refund money column
refund_df
```

```python
# Calculate total profit from sales (Quantity × Price)
sales_df['Profit'] = sales_df['Quantity'] * sales_df['Price']
```

```python
# Display sales data with profit column
sales_df
```

```python
# Calculate total refund amount (Quantity × Price)
refund_df['Refund Money'] = refund_df['Quantity'] * refund_df['Price']
refund_df
```

```python
# Extract discount/promotional transactions (price = 0)
discount_df = df_golden[df_golden['Price'] == 0]
discount_df
```

```python
# Display sample of discount transactions
discount_df.sample(6)
```

```python
# Calculate average price per product (StockCode) and map to discount items
# shows what the typical price should have been for discounted items
avg_prices = df_golden.groupby('StockCode')['Price'].mean().round(2)
discount_df['Average Price'] = discount_df['StockCode'].map(avg_prices)
discount_df[discount_df['Average Price'] == 'M']
```

```python
# Display discount data with average price mapping
discount_df
```

## Upload to SSMS


```python
# ========================================
# DATABASE LAYER: Upload to SQL Server
# ========================================

from sqlalchemy import create_engine
import urllib
```

```python
# Configure SQL Server connection details
server = r'DESKTOP-8N5RPBA\MSSQLSERVER01'  # Local SQL Server instance
database = 'Retail'  # Target database
```

```python
# Create SQLAlchemy engine with Windows authentication
connection_string = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'
params = urllib.parse.quote_plus(connection_string)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
# Write Sales data to SQL Server (replace if exists)
sales_df.to_sql('Sales', con=engine, schema='Golden', if_exists='replace', index=False)
```

```python
# Write Refund data to SQL Server (replace if exists)
refund_df.to_sql('Refund', con=engine, schema='Golden', if_exists='replace', index=False)
```

```python
# Write Discount data to SQL Server (replace if exists)
discount_df.to_sql('Discounts', con=engine, schema='Golden', if_exists='replace', index=False)
```
